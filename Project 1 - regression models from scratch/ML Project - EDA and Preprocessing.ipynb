```python
months = [col.split()[0] for col in min_temp_cols]

#flip min and max temperatures if they are the wrong way around
for ix,row in df.iterrows():
    for month in months:
        if row[month + ' mean min temperature'] > row[month + ' mean max temperature']:
            new_max_temp = row[month + ' mean min temperature']
            new_min_temp = row[month + ' mean max temperature']
            df.at[ix, month + ' mean min temperature'] = new_min_temp
            df.at[ix, month + ' mean max temperature'] = new_max_temp

#remove rows with extreme max temperatures
df = df[df['oct mean max temperature']<85]
df = df[df['mar mean max temperature']<80]
df.at[df['may rain']<0,'may rain']=0
```

## Split data

```python
fig,axs = plt.subplots(1,2,sharey=True,figsize=[15,5])
axs[0].hist(df_c['oct rain'],bins=20)
axs[0].set_title('Oct Rain (without transform)')
axs[1].hist(df_l['oct rain'],bins=20)
axs[1].set_title('Oct Rain after log transform')
plt.savefig("figures\\transformhist.png")
plt.show()
```

```python
def ttv_split(data,trainp,testp,rs=0):
    
    x_raw = data.drop('yield',axis=1)
    y_raw = data['yield']
    
    total_count = data.shape[0]
    valp = 1 - testp - trainp

    trains = int(total_count * trainp)
    vals = int(total_count * valp) + trains

    np.random.seed(rs)
    shuffle = np.random.permutation(total_count)

    x = x_raw.iloc[shuffle]
    y = y_raw.iloc[shuffle]

    x_train_unnorm = x.iloc[:trains]
    y_train_unnorm = y.iloc[:trains]

    x_val_unnorm = x.iloc[trains:vals]
    y_val_unnorm = y.iloc[trains:vals]

    x_test_unnorm = x.iloc[vals:]
    y_test_unnorm = y.iloc[vals:]
    
    print('Training set size:', x_train_unnorm.shape[0])
    print('Validation set size:', x_val_unnorm.shape[0])
    print('Test set size:', x_test_unnorm.shape[0])
    
    return x_train_unnorm, y_train_unnorm, x_val_unnorm, y_val_unnorm, x_test_unnorm, y_test_unnorm
```

```python
x_train_unnorm, y_train, x_val_unnorm, y_val, x_test_unnorm, y_test = ttv_split(df_l,0.7,0.15)
```

## Normalise Data

```python
def normalise_fit(x):
    mu = x.mean()
    std = x.std()
    
    x_normalised = (x - mu)/std
    
    return x_normalised, mu, std

def normalise_transform(x, mu, std):
    return (x - mu ) / std


x_train_norm, mu, std = normalise_fit(x_train_unnorm)
x_val_norm = normalise_transform(x_val_unnorm, mu, std)
x_test_norm = normalise_transform(x_test_unnorm, mu, std)
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_viz = PCA()
x_pca = pca_viz.fit_transform(x_train_norm)
```

```python
expl_var = pca_viz.explained_variance_ratio_ * 100
expl_var_cum = np.cumsum(expl_var)
```

```python
sns.set_context("talk")
plt.figure(figsize=[10,5])
plt.plot(list(range(1,len(expl_var_cum)+1)),expl_var_cum, marker='o', linestyle='-',markersize=8,linewidth=2)
plt.axhline(99,c='r',linestyle='--')
plt.axvline(20,c='r',linestyle='--',linewidth=1)
plt.yticks(list(range(40,101,10)))
plt.xlabel('# PCA Components')
plt.ylabel('Cumulative % variance')
plt.grid()
plt.savefig(r"figures\pcaplot.png")
plt.show()
```
